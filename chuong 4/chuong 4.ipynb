# bài 4.20

```python
euro12 = pd.read_csv("euro12.csv")
print('giá trị cột goals')
print(euro12['Goals'])
print("Số đội tham gia Euro 2012:", euro12['Team'].nunique())
print('Thông tin dữ liệu')
euro12.info()
print('Tạo DataFrame discipline')
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print(discipline)
print('Sắp xếp giảm dần theo Red Cards, Yellow Cards')
print(discipline.sort_values(
    by=['Red Cards', 'Yellow Cards'],
    ascending=False
))
print('Trung bình Yellow Cards:',euro12['Yellow Cards'].mean())
print('Các đội ghi hơn 6 bàn')
print(euro12[euro12['Goals'] > 6])
print('Các đội có tên bắt đầu bằng "G"')
print(euro12[euro12['Team'].str.startswith('G')])
print('7 cột đầu')
print(euro12.iloc[:, :7])
print('Tất cả cột trừ 3 cột cuối')
print(euro12.iloc[:, :-3])
print('Các cột yêu cầu')
print(euro12[['Team', 'Goals', 'Shooting Accuracy', 'Yellow Cards', 'Red Cards']])
print('Team & Shooting Accuracy của England, Italy, Russia')
print(euro12.loc[
    euro12['Team'].isin(['England', 'Italy', 'Russia']),
    ['Team', 'Shooting Accuracy']
])
```

# bài 4.21

```python
drink = pd.read_csv("drinks.csv", index_col=0)
drink.info()
drink.shape
drink.columns

print('Trung bình beer theo châu lục')
print(drink.groupby('continent')['beer_servings'].mean())

print('Thống kê wine theo châu lục')
print(drink.groupby('continent')['wine_servings'].describe())

print('Trung bình bia & rượu theo châu lục')
print(drink.groupby('continent').mean())

print('Trung vị bia & rượu theo châu lục')
print(drink.groupby('continent').median())

print('Spirit_servings: mean, max, min')
print(drink.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min']))

print('5 nước uống bia nhiều nhất / ít nhất')
print(drink.sort_values('beer_servings', ascending=False).head())
print(drink.sort_values('beer_servings').head())

```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
movies.info()
ratings.info()
movies.isnull().sum()
ratings.isnull().sum()
movies.dropna(inplace=True)
ratings = ratings[ratings['rating'] > 0]
movie_data = ratings.merge(movies, on='movieId')
movie_data.head()
movie_data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```
