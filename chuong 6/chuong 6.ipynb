# bai 6.1

```python
import threading

def print_thread_name():
    print("Thread name:", threading.current_thread().name)

threads = []
for i in range(5):
    t = threading.Thread(target=print_thread_name, name=f"Thread-{i+1}")
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

# bai 6.2

```python
import threading
import urllib.request

def download(url, filename):
    urllib.request.urlretrieve(url, filename)
    print(f"Downloaded {filename}")

files = {
    "https://www.example.com": "file1.html",
    "https://www.python.org": "file2.html"
}

threads = []
for url, name in files.items():
    t = threading.Thread(target=download, args=(url, name))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

# bai 6.3

```python
def even_numbers():
    for i in range(30, 51):
        if i % 2 == 0:
            print("Even:", i)

def odd_numbers():
    for i in range(30, 51):
        if i % 2 != 0:
            print("Odd:", i)

t1 = threading.Thread(target=even_numbers)
t2 = threading.Thread(target=odd_numbers)

t1.start()
t2.start()

t1.join()
t2.join()
```

# bai 6.4

```python
import math

def factorial(n):
    print(f"{n}! = {math.factorial(n)}")

numbers = [5, 7, 10]

threads = []
for n in numbers:
    t = threading.Thread(target=factorial, args=(n,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

# bai 6.5

```python
import requests

def fetch(url):
    r = requests.get(url)
    print(url, "->", r.status_code)

urls = [
    "https://www.google.com",
    "https://www.yahoo.com",
    "https://www.facebook.com"
]

threads = []
for url in urls:
    t = threading.Thread(target=fetch, args=(url,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

# bai 6.6

```python
def print_even(n):
    for i in range(0, n + 1, 2):
        print("Even:", i)

def print_odd(n):
    for i in range(1, n + 1, 2):
        print("Odd:", i)

t1 = threading.Thread(target=print_even, args=(20,))
t2 = threading.Thread(target=print_odd, args=(20,))

t1.start()
t2.start()

t1.join()
t2.join()
```

# bai 6.7

```python
import time
from datetime import datetime

def worker(name, delay):
    print(f"Starting {name}")
    for _ in range(3):
        time.sleep(delay)
        print(f"{name}: Web {datetime.now()}")
    print(f"Exiting {name}")

print("Starting Main Thread")

threads = [
    threading.Thread(target=worker, args=("Google", 1)),
    threading.Thread(target=worker, args=("Yahoo", 2)),
    threading.Thread(target=worker, args=("Facebook", 3))
]

for t in threads:
    t.start()

for t in threads:
    t.join()

print("Exiting Main Thread")
```

# bai 6.8

```python
import random

def partial_sum(lst, result, index):
    result[index] = sum(lst)

n = 40
data = [random.randint(0, 10) for _ in range(n)]

parts = [data[:n//3], data[n//3:2*n//3], data[2*n//3:]]
result = [0] * 3

threads = []
for i in range(3):
    t = threading.Thread(target=partial_sum, args=(parts[i], result, i))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("List:", data)
print("Total sum:", sum(result))
```

# bai 6.10

```python
def find_max(lst, result, index):
    result[index] = max(lst)

n = 30
data = [random.randint(0, 100) for _ in range(n)]

parts = [data[:10], data[10:20], data[20:]]
result = [0] * 3

threads = []
for i in range(3):
    t = threading.Thread(target=find_max, args=(parts[i], result, i))
    threads.append(t)
    t.start()

for t in threads:
    t.join()

print("List:", data)
print("Max value:", max(result))
```
